```python
import pandas as pd
import numpy as np
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
data
```

```python
a = pd.Series(data)
a
```

```python
mean = a.mean()
mean
```

```python
std = a.std()
std
```

```python
from scipy import stats
```

```python
n = len(data)

# Degrees of freedom
df = n - 1

# t-critical value for 99% confidence and it is nothing but inversly proportional to the cdf(cumulative distribution function)
t_critical = stats.t.ppf(1 - 0.005, df)

# Margin of error
margin_error = t_critical * (std / np.sqrt(n))

# Confidence interval
lower_bound = mean - margin_error
upper_bound = mean + margin_error

print(f"99% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
from scipy.stats import norm
```

```python
n = len(data)
mean = a.mean()
# Z-critical value for 99% confidence
z_critical = norm.ppf(1 - 0.005)

# Margin of error
margin_error = z_critical * (n / np.sqrt(n))

# Confidence interval
lower_bound = mean - margin_error
upper_bound = mean + margin_error

print(f"99% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})"
```
