```python
import pandas as pd
import numpy as np
```

```python

from scipy.stats import norm

# Given values
sample_mean = 3050
sample_size = 25
units_mean = 600
units_std_dev = 25

# Theoretical weekly cost model: W = 1000 + 5X
theoretical_mean = 1000 + 5 * units_mean  # = 4000
population_std_dev = 5 * units_std_dev    # = 125

# Calculate z-test statistic
z = (sample_mean - theoretical_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate p-value (one-tailed test)
p_value = 1 - norm.cdf(z)

# Decision threshold
alpha = 0.05
decision = "Reject H₀" if p_value < alpha else "Fail to reject H₀"

# Display results using pandas
results = pd.DataFrame({
    "Sample Mean": [sample_mean],
    "Theoretical Mean": [theoretical_mean],
    "Z-Statistic": [z],
    "P-Value": [p_value],
    "Decision": [decision]
})

print(results)
```

```python
"""The hypothesis test shows no statistical evidence of an increase in weekly operating costs. The restaurant owners' concerns, while valid operationally, are not supported by this sample data."""
```
